```python
#https://github.com/abishekk92/potter
fname = 'book_1.txt'
books = {}
with open(fname) as f:
    if fname.endswith('.txt'): #Each file name is just the given year, so we can use this to create the column. Getting rid of .csv
        title = fname[:-4]
    text = f.read()
    books[title] = text 
books.keys()
```

## Frequency Counters

### Base RE Frequency Counter

```python
b1_words = re.split(r'\W+', books['book_1']) #Basic text cleaning
```

```python
b1_FD = nltk.FreqDist(b1_words)
for x in b1_FD:
    print(x ,b1_FD[x])
```

### NLTK Frequency List 
Again, based on Prof. Cavar's notebooks and https://machinelearningmastery.com/clean-text-machine-learning-python/

```python
tokens_nltk = nltk.word_tokenize(books['book_1'])
tokens_nltk = [w.lower() for w in tokens_nltk]

table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in tokens_nltk]
words = [word for word in stripped if word.isalpha()]
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
words = [w for w in words if not w in stop_words]

nltk_freq = Counter(words)
nltk_freq.most_common(50)
```

### SpaCy Frequency List

```python
nlp = spacy.load('en')
hp1 = nlp(books['book_1'])#If you run into trouble, spaCy Only takes unicode 
```

```python
# If having trouble loading your english model, try 
# python -m spacy download en
# on the command line
tokens = [token.lower_ for token in hp1 if token.is_alpha and not token.is_stop]
spacy_freq = Counter(tokens)
spacy_freq.most_common(50)
```

As you can see, both result very similar results without much difference in work. However, it is interesting to look at which stopwords are/are not included. Both stop word lists can be modified, but it is worth being away or the differences between the default lists.

## Named Entity Recognition
Based heavily off of tutorial given https://towardsdatascience.com/named-entity-recognition-with-nltk-and-spacy-8c4a7d88e7da

```python
labels = [x.label_ for x in hp1.ents]
Counter(labels)
```

```python
sentences = [x for x in hp1.sents]
```

```python
from spacy import displacy
displacy.render(nlp(str(sentences[33:50])), jupyter=True, style="ent")
```

This is a great example of the power of spaCy and how easy it can be to have a visual NER result. You can also see, however, where more fine tuning would be necessary; It recognizes Dursley as a person, but fails to pick up that Dudley is also a person.
